```python
import requests
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

## Premier league

```python
years = [y for y in range(2001,2024)]

years_url = []

premier_seasons_teams = []
for year in years:
    url = "http://sports.core.api.espn.com/v2/sports/soccer/leagues/eng.1/seasons/"+str(year)+"/teams"
    resp = requests.get(url)
    data = resp.json()
    # years_url.append(data['$ref'])
    print(year)
    # print(data["items"])
    for item in data["items"]:
        
        url = item["$ref"]
        resp = requests.get(url)
        team_data = resp.json()
        # print(team_data["name"])
        team_info = {'season':year, 'name':team_data["name"],'id':team_data["id"] } 
        premier_seasons_teams.append(team_info)
```

```python
import json
print(premier_seasons_teams)
file_path = 'data/premier_seasons_teams.json'

# Open the file in write mode ('w') and write the JSON data
with open(file_path, 'w') as file:
    json.dump(premier_seasons_teams, file)
```
